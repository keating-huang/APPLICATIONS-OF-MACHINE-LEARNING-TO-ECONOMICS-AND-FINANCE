### Seaborn
seaborn為python繪圖函式庫，以matplotlib為基礎封裝了許多實用的統計圖表，相較於matplotlib，seaborn的預設圖形更為好看，語法更為簡潔，並且良好的支援了pandas函式庫。

#### 優點：
`畫圖指令非常簡單

`圖案好看

#### 缺點：
`雖然99%的使用上都沒問題，但有些進階的圖案很難畫或是沒辦法畫（可用matplot互補）

```python
sns.catplot(x='sex',y='total_bill',data=tips, kind="bar")
```

```python
sns.countplot(x='sex',data=tips) #算人数
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="coolwarm")
#hue 分别画 ‘smoker’有和无
```

```python
# Matrix form for correlation data
tips.corr() #查看相关系数。相关性 -1~1
```

```python
sns.heatmap(tips.corr()) #热图
```

```python
sns.heatmap(tips.corr(),cmap='coolwarm',annot=True) #annot显示数字
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter') #jointplot
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

### Matplotlib
#### 優點：
相較於其他視覺化套件，matplot算是最歷史悠久，因此有很多的教學文章或是範例可參考；

畫圖功能最齊全，基本上沒什麼圖表畫不出來的

#### 缺點:
圖表不好看(新版的matplot其實也算好看，尤其在style功能出來之後可以自行切換圖表的風格)；

畫圖指令複雜

```python
import matplotlib.pyplot as plt

x=[5,4,3,2,1]
y=[1,2,3,4,5]

plt.plot(x,y,color='blue')
plt.show()
```

```python
plt.plot(x,y,color='#0000ff') #alternative
plt.show()
```

```python
plt.plot(x,y,'o')
```

```python
plt.plot(x,y,color='g',linestyle='dashed',markeredgecolor='r',marker='+')
```

```python
x=np.array([5,4,3,2,1])
y=pow(x,3)
plt.axis([1,6,1,150])
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,label='y = x ** 3')
plt.legend()
```

```python
plt.plot(x,y,label='y = x ** 3')
plt.legend(loc=4,shadow=True)
```

```python
z=pow(x,4)
w=pow(x,5)
plt.figure(figsize=(8,6))
plt.plot(x,y)
plt.plot(x,z)
plt.plot(x,w)
plt.title("The result")
plt.text(4.5,200,'x**3')
plt.text(4,400,'x**4')
plt.text(3.5,900,'x**5')
plt.show()
```

```python
df=sns.load_dataset('titanic')
df.head(10)
```

```python
df['age'].plot.hist(bins=30)
```

```python
#製作survived和sex的交叉表
df.groupby(['survived','sex']).size().unstack(1)
```

```python
#將結果製成圖
df.groupby(['survived','sex']).size().unstack(1).plot(kind='bar')
```

```python
#製作survived和age的圖。將依存活與否分成兩組，再將兩組的資料分別用直方圖來呈現。畫在同一格表格上
df.groupby(['survived','age']).size().unstack(1)
```

```python
df.groupby('survived')['age'].plot(kind='hist', bins=30, alpha=0.7, legend=True)
```

```python
#用箱型圖來呈現
df.groupby(['survived','age']).size().unstack(0).plot(kind='box')
```

```python
#用箱型圖來呈現（ans)
df.boxplot(column='age',by='sex')
```
